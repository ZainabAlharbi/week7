```python
import keras
import numpy as np 
import pandas as pd
```

```python
df = pd.read_excel('/content/drive/MyDrive/ML_MODEL/credit-data-training.xlsx')
```

```python
df.shape
```

```python
#X = df.drop(columns=['Credit-Application-Result','Value-Savings-Stocks','Guarantors'])
#R = df[['Credit-Application-Result']]
X = pd.get_dummies(df.drop(['Credit-Application-Result','Value-Savings-Stocks','Guarantors','Duration-in-Current-address'], axis=1))
y = df['Credit-Application-Result'].apply(lambda x: 1 if x=='Creditworthy' else 0)
X = X.dropna()
```

```python
X
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(17, activation='relu', input_dim=len(X.columns)))
model.add(keras.layers.Dense(17, activation='relu'))
model.add(keras.layers.Dense(1))


model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
model.fit(X,y,epochs=100)
```
